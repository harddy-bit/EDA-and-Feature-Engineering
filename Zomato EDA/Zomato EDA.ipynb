## Methods of finding Missing Values
1. Normal isnull
2. List comprehension method
3. Heatmap

```python
df.shape
```

```python
#1 - Normal

df.isnull().sum()
```

```python
##2 - List Comprehension

[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# 3 - HeatMap - Here not able to see in chart as missing value(9) is less in number

sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
df_final = pd.merge(df,df_country,on='Country Code',how='left')
df_final.head(2)
```

```python
## To check datatypes

df_final.dtypes
```

```python
df_final.columns
```

```python
df_final.Country.value_counts()
```

```python
country_names = df_final.Country.value_counts().index
country_names
```

```python
country_values = df_final.Country.value_counts().values
country_values
```

```python
## Pie Chart - Top 3 Countries that uses Zomato

plt.pie(country_values[:3],labels=country_names[:3])
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

# Observation - Zomato Maximum Records or transcation are from India. After that USA and then United Kingdom



```python
df_final.columns
```

```python
df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
ratings = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

## Observation from above 
1. When ratings are between 4.5-4.9 ---> Excellent
2. When ratings are between 4.0-4.4 ---> Very Good
3. When ratings are between 3.5-3.9 ---> Good
4. When ratings are between 2.5-3.4 ---> Average
5. When ratings are between 1.8-2.4 ---> Poor
6. When ratings are 0 ---> Given by many people

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) ## To make diagram bigger
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
## Plot as count plot

sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
